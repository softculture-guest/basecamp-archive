```python
import os
from PIL import Image

folder = "/Users/noideaatall/_trash/blog/blog"
os.chdir(folder)
```

```python
wrong = 0

for filename in os.listdir(os.curdir):
    #print(filename)
    try:
        image = Image.open(filename)
    except:
        continue
        
    width = image.size[0]
    height = image.size[1]
    mb = round(os.path.getsize(filename)/1048576.0, 2)
    res = (width*height)/1000000
    compression = round(mb/res,5)
    
    if width > 2000 or compression > 0.5:
        wrong += 1
        print(filename)
        print(" {}×{}, {} Mb, {}".format(width, height, mb, compression))
        
print("{} wrong images found".format(wrong))
```
